```python
import numpy as np
import pandas as pd
import os
#%config IPCompleter.greedy=True
```

```python
filepath =str((os.getcwd())) + "/data/quest_analytics_colocation_data.csv"
practice_data = pd.read_csv(filepath)
practice_data.head(5)

# 1 - Identify co-located practices in the data set
practice_data['co_location_counts'] = practice_data.groupby(['practice_street_line_1','organization_name'])['practice_street_line_1'].transform('count')
# Filtering the Practice info to display co-located practices later
practice_info = practice_data.filter(['practice_street_line_1','practice_street_line_2','practice_city','practice_state','practice_phone','practice_fax','co_location_counts'], axis=1)

# Only get the practices that are co-located
practice_colocation_info = practice_info[(practice_info.co_location_counts >= 2)]
# Show the Co-Located Practices
practice_colocation_info.reset_index(drop=True)
```

```python
# 2 - Split up co-located practices, and determine which providers belong with which practices
practices_providers = practice_data.filter(['provider_first_name','provider_last_name','organization_name','co_location_counts'])
# Dropping duplicate records
colocation_practices = practices_providers[(practices_providers.co_location_counts >= 2)].drop(['co_location_counts'],axis=1).drop_duplicates()
colocation_practices.rename(columns={'organization_name':'colocated_practice'}, inplace=True)
# Show the providers and their co_located practices
colocation_practices.set_index('colocated_practice')
```
